```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input
from tensorflow.keras import layers, models, metrics, callbacks
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras import optimizers
```


DATA PREPARATION



```python
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=30,
    width_shift_range=0.25,
    height_shift_range=0.25,
    shear_range=0.2,
    zoom_range=0.3,
    horizontal_flip=True,
    fill_mode='constant'
)

val_test_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

# Data generators with explicit RGB conversion
train_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/dataset/train_pdr',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    color_mode='rgb',
    shuffle=True
)

val_generator = val_test_datagen.flow_from_directory(
    '/content/drive/MyDrive/dataset/val_pdr',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    color_mode='rgb'
)
```

MODEL ARCHITECTURE

```python
def build_attention_efficientnet():
    # Base model
    base_model = EfficientNetB0(
        weights='imagenet',
        include_top=False,
        input_shape=(224, 224, 3)
    )

    # Enhanced attention mechanism
    def channel_attention(input_tensor):
        channel_axis = -1
        filters = input_tensor.shape[channel_axis]

        shared_layer = layers.Dense(filters//8, activation='relu')

        avg_pool = layers.GlobalAveragePooling2D()(input_tensor)
        avg_pool = layers.Reshape((1, 1, filters))(avg_pool)
        avg_pool = shared_layer(avg_pool)

        max_pool = layers.GlobalMaxPooling2D()(input_tensor)
        max_pool = layers.Reshape((1, 1, filters))(max_pool)
        max_pool = shared_layer(max_pool)

        attention = layers.Add()([avg_pool, max_pool])
        attention = layers.Conv2D(filters, 1, activation='sigmoid')(attention)

        return layers.Multiply()([input_tensor, attention])

    # Model construction
    x = base_model.output
    x = channel_attention(x)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(512, activation='relu')(x)
    x = layers.Dropout(0.6)(x)
    outputs = layers.Dense(5, activation='softmax')(x)

    return models.Model(inputs=base_model.input, outputs=outputs)

model = build_attention_efficientnet()
```


TRAINING SETUP

```python
# Class weights for severe imbalance
class_weights = {
    0: 1.5,  # Blurry
    1: 1.0,  # Normal
    2: 3.0,  # Moderate
    3: 5.0,  # Severe
    4: 8.0   # Malignant
}

# Callbacks for medical use case
callbacks = [
    callbacks.ModelCheckpoint(
        'best_model.keras',
        monitor='val_auc',
        mode='max',
        save_best_only=True,
        save_weights_only=False
    ),
    callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=2,
        min_lr=1e-7
    ),
    callbacks.EarlyStopping(
        monitor='val_auc',
        patience=8,
        mode='max',
        restore_best_weights=True
    )
]

# ======================
# TRAINING PROCESS
# ======================

# Phase 1: Frozen base training
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=[
        'accuracy',
        metrics.AUC(name='auc', multi_label=True),
        metrics.Recall(name='sensitivity')
    ]
)

history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=15,
    validation_data=val_generator,
    validation_steps=len(val_generator),
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=1
)

# Phase 2: Fine-tuning
model.trainable = True
for layer in model.layers[:150]:
    layer.trainable = False

model.compile(
    optimizer=optimizers.Adam(1e-5),
    loss='categorical_crossentropy',
    metrics=[
        'accuracy',
        metrics.AUC(name='auc', multi_label=True),
        metrics.Recall(name='sensitivity')
    ]
)

history_fine = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=25,
    initial_epoch=history.epoch[-1],
    validation_data=val_generator,
    validation_steps=len(val_generator),
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=1
)
```

EVALUATION

```python
test_generator = val_test_datagen.flow_from_directory(
    '/content/drive/MyDrive/dataset/test_pdr',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    color_mode='rgb',
    shuffle=False
)

# Detailed evaluation
test_metrics = model.evaluate(test_generator, return_dict=True)
print("\nFinal Test Performance:")
for k, v in test_metrics.items():
    print(f"{k}: {v:.4f}")

# Enhanced confusion matrix
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

y_true = test_generator.classes
y_pred = model.predict(test_generator).argmax(axis=1)
class_names = list(test_generator.class_indices.keys())

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

print("\nConfusion Matrix:")
ConfusionMatrixDisplay.from_predictions(
    y_true, y_pred,
    display_labels=class_names,
    cmap='Blues',
    normalize='true'
)
plt.show()
```

```python
# New cell for Gradio web interface
import gradio as gr
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.efficientnet import preprocess_input

# Load saved model
model = load_model('/content/drive/MyDrive/model/best_model.keras')

# Class configuration
class_names = ['Blurry', 'Normal', 'Moderate', 'Severe', 'Malignant']

def predict_image(img):
    # Preprocess image for EfficientNet
    img = img.resize((224, 224))
    img_array = np.array(img)
    img_array = preprocess_input(img_array)
    img_array = np.expand_dims(img_array, axis=0)

    # Make prediction
    predictions = model.predict(img_array)[0]

    # Convert to percentage dictionary
    return {class_names[i]: float(predictions[i]) for i in range(5)}

# Create Gradio interface
interface = gr.Interface(
    fn=predict_image,
    inputs=gr.Image(type="pil", label="Upload Fundus Image"),
    outputs=gr.Label(num_top_classes=5, label="Prediction Probabilities"),
    title="Diabetic Retinopathy Classification",
    description="Upload a retinal fundus image to classify severity (Blurry, Normal, Moderate, Severe, Malignant)"
)

# Launch the web app
interface.launch(share=True)
```

```python
from IPython.display import display, HTML
from IPython.html import widgets
from PIL import Image
import numpy as np

# Load best saved model
model = load_model('best_model.keras')

# Class configuration (matches training setup)
class_names = {
    0: 'Blurry',
    1: 'Normal',
    2: 'Moderate',
    3: 'Severe',
    4: 'Malignant'
}

class_weights = {
    0: 1.5,  # Blurry
    1: 1.0,  # Normal
    2: 3.0,  # Moderate
    3: 5.0,  # Severe
    4: 8.0   # Malignant
}

# Create upload widget
upload = widgets.FileUpload(accept='.jpg,.jpeg,.png', multiple=False)
display(upload)

def process_uploaded_image(_):
    if upload.data:
        try:
            # Load and display image
            img = Image.open(upload.data[0]).convert('RGB')
            display(img.resize((224, 224)))

            # Preprocess for EfficientNet
            img_array = np.array(img.resize((224, 224)))
            img_array = preprocess_input(img_array)
            img_array = np.expand_dims(img_array, axis=0)

            # Make prediction
            predictions = model.predict(img_array)
            predicted_class = np.argmax(predictions[0])
            confidence = np.max(predictions[0])

            # Display clinical results
            html = f"""
            <style>.result-box {{border: 2px solid #e0e0e0; padding: 20px; margin: 10px;}}</style>
            <div class="result-box">
                <h3>Clinical Prediction Results</h3>
                <p><b>Class:</b> {class_names[predicted_class]}</p>
                <p><b>Confidence:</b> {confidence:.2%}</p>
                <p><b>Class Weight:</b> {class_weights[predicted_class]}</p>
                <p><b>Clinical Note:</b> {['Consider retaking image',
                                         'No significant pathology',
                                         'Moderate changes observed',
                                         'Urgent specialist review needed',
                                         'Immediate oncology referral'][predicted_class]}</p>
            </div>
            """
            display(HTML(html))

        except Exception as e:
            print(f"Error: {str(e)}")

# Set up prediction trigger
upload.observe(process_uploaded_image, names='data')
```
